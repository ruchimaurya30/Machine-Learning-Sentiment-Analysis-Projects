In this pivotal chapter of our sentiment analysis journey, you've harnessed NLTK's SentimentIntensityAnalyzer to unveil the intricate emotions within Amazon product reviews.
```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()
# --- WRITE YOUR CODE FOR TASK 4 ---
# define a function to compute sentiment scores
def sentiment_scores(Text):
    scores = sentiments.polarity_scores(Text)
    return scores['pos'], scores['neg'], scores['neu']

# apply the function to each row in the text column and create new column
df[['Positive', 'Negative', 'Neutral']] = df['Text'].apply(lambda x : pd.Series(sentiment_scores(x)))
```
